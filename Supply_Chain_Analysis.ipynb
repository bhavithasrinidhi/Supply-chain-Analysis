## Step 2: Revenue Analysis by Product Type
Which product category generates the most revenue?

```python
# Revenue by Product Type
print("Total Revenue by Product Type:")
print(df.groupby('Product type')['Revenue generated'].sum().sort_values(ascending=False))

# Visualization
plt.figure(figsize=(8,5))
df.groupby('Product type')['Revenue generated'].sum().sort_values(ascending=False).plot(kind='bar', color='#2E75B6')
plt.title('Total Revenue by Product Type', fontsize=14, fontweight='bold')
plt.xlabel('Product Type')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Insight 1 — Revenue by Product Type
- Skincare leads with $241,628 (41.6% of total revenue)
- Haircare second with $174,455 (30.1%)
- Cosmetics lowest with $161,521 (27.8%)
- Skincare generates ~50% more revenue than Cosmetics
- Business should prioritise skincare inventory and marketing

## Step 3: Quality Analysis — Defect Rate by Product Type
Which product type has the most quality issues?

```python
# Defect rate by product type
print("Average Defect Rate by Product Type:")
print(df.groupby('Product type')['Defect rates'].mean().sort_values(ascending=False))

plt.figure(figsize=(8,5))
df.groupby('Product type')['Defect rates'].mean().sort_values(ascending=False).plot(kind='bar', color='#C00000')
plt.title('Average Defect Rate by Product Type', fontsize=14, fontweight='bold')
plt.xlabel('Product Type')
plt.ylabel('Defect Rate (%)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Insight 2 — Defect Rate by Product Type
- Haircare has highest defect rate at 2.48%
- Skincare follows closely at 2.33%
- Cosmetics has best quality with only 1.92% defects
- Haircare defect rate is 29% higher than Cosmetics
- Quality improvement should prioritise Haircare production

## Step 4: Inventory Analysis — Stock vs Sales
Are we holding too much or too little inventory?

```python
# Stock levels vs Products sold by category
inventory = df.groupby('Product type')[['Stock levels','Number of products sold']].mean()
print(inventory)

inventory.plot(kind='bar', figsize=(8,5), color=['#2E75B6','#70AD47'])
plt.title('Average Stock Levels vs Products Sold', fontsize=14, fontweight='bold')
plt.xlabel('Product Type')
plt.ylabel('Units')
plt.xticks(rotation=0)
plt.legend(['Stock Levels','Products Sold'])
plt.tight_layout()
plt.show()
```

### Insight 3 — Stock Levels vs Sales (Critical!!)
- Skincare sells the most (518 units avg) but holds least stock (40 units avg)
- Skincare stock-to-sales ratio is only 7.76% - lowest of all categories
- Cosmetics and Haircare have healthier buffers (12-13%)
- Skincare is at HIGH RISK of stockouts given low buffer + highest demand
- Recommendation: Increase skincare safety stock levels urgently

## Step 5: Logistics Analysis — Shipping Costs by Carrier
Which shipping carrier offers the best cost efficiency?

```python
# Shipping costs by carrier
print("Average Shipping Cost by Carrier:")
print(df.groupby('Shipping carriers')['Shipping costs'].mean().sort_values())

print("\nAverage Shipping Time by Carrier:")
print(df.groupby('Shipping carriers')['Shipping times'].mean().sort_values())

plt.figure(figsize=(8,5))
df.groupby('Shipping carriers')['Shipping costs'].mean().sort_values().plot(kind='bar', color='#2E75B6')
plt.title('Average Shipping Cost by Carrier', fontsize=14, fontweight='bold')
plt.xlabel('Shipping Carrier')
plt.ylabel('Average Cost ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Insight 4 — Shipping Carrier Performance
- Carrier B is the best choice - cheapest ($5.51) AND fastest (5.30 days)
- Carrier A is slowest (6.14 days) despite similar cost to Carrier B
- Carrier C is most expensive ($5.60) with mediocre speed
- Carrier A takes 16% longer than Carrier B for almost same price
- Recommendation: Shift more shipments to Carrier B for cost and speed efficiency

## Step 6: Transportation Mode Analysis
Which mode of transport is most cost-efficient?

```python
# Costs by transportation mode
print("Average Costs by Transportation Mode:")
print(df.groupby('Transportation modes')['Costs'].mean().sort_values())

plt.figure(figsize=(8,5))
df.groupby('Transportation modes')['Costs'].mean().sort_values().plot(kind='bar', color='#70AD47')
plt.title('Average Costs by Transportation Mode', fontsize=14, fontweight='bold')
plt.xlabel('Transportation Mode')
plt.ylabel('Average Cost ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Insight 5 — Transportation Mode Costs
- Sea transport is cheapest at $417.82 average cost
- Air transport is most expensive at $561.71 (34% higher than Sea)
- Rail and Road fall in between with similar costs (~$540-553)
- Recommendation: Use Sea transport for non-urgent shipments to reduce costs
- Reserve Air transport only for time-critical/urgent deliveries
